```python
import requests
import csv
```

```python
baseurl = 'https://api.nasa.gov/neo/rest/v1/neo/browse?'
pagenumber = 'page=1'
authenticationkey = '&size=20&api_key=DEMO_KEY'

def main_request(baseurl, pragenumber, authenticationkey):
    r = requests.get(baseurl + pagenumber + authenticationkey)
    return r.json()

def nextPage(pagenumber):
    base = 'page='
    tmp = int(pagenumber[-1])
    tmp += 1
    pagenumber = base + str(tmp)
    return pagenumber

cleanedList = []
i = 0
dataList = main_request(baseurl,pagenumber,authenticationkey)

while i < 4:

    for item in dataList['near_earth_objects']:
        #print(f'{item['name']}, is hazardous: {item['is_potentially_hazardous_asteroid']}')
        #print(f'mid-range estimate of asteriod\'s size (in miles): {(item['estimated_diameter']['miles']['estimated_diameter_min']+item['estimated_diameter']['miles']['estimated_diameter_max'])/2}')
        #print(f'frequency of near misses (from the years 1900-2200): {len(item['close_approach_data'])}')
        tmp = [item['name'], (item['estimated_diameter']['miles']['estimated_diameter_min']+item['estimated_diameter']['miles']['estimated_diameter_max'])/2, item['is_potentially_hazardous_asteroid'], len(item['close_approach_data'])]
        cleanedList.append(tmp)

    pagenumber = nextPage(pagenumber)
    dataList = main_request(baseurl,pagenumber,authenticationkey)
    i += 1

cleanedList

```

```python
header = ['Name of Asteriod', 'Mid-Range Size Estimate (in miles)', 'Is Potentially Hazardous?', 'Frequency of Near Misses']

with open('projectData.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(header)
    writer.writerows(cleanedList)
```
